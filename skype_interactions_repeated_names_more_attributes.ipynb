```python
import random
import numpy as np
import pandas as pd
import sys
from tqdm import tqdm
```

```python
def generate_interaction_mins(n,target):
    x = np.random.randint(0, target, size=(n,))
    while sum(x) != target:
        x = np.random.randint(0, target, size=(n,))
    return x
```

```python
def skype_interactions(n):
    df = pd.DataFrame(columns = ['organizer' , 'call_duration_mins', 'participants' ,'indiv_interaction_mins',
                                 'interaction','meeting_modes','meeting_types','host_locations','recorded_session'])
    
    first_name = ['Vaishali','Aruna','Prasanna','Avinash','Harini','Divya','Surya']
    second_name=['Ravindranath','Prasad','Raj','Prakash']
    
    levels = ['High','Low','Medium']
    meeting_mode = ['in_office','online','hybrid']
    meeting_type=['project_discussion','employee_engagement','trainings','others']
    host_location=['India','France','Romania','Remote']
    
    for i in range(n):
        organiser=random.choice(first_name)+' '+random.choice(second_name)
        call_duration_mins=random.randint(5,180)
        group=[",".join(random.choice(first_name)+' '+random.choice(second_name) for _ in range(random.randint(2,10)))]
        participants=(set([organiser]) ^ set(group))
        indiv_interaction_mins=generate_interaction_mins(len(participants)+1,call_duration_mins)
        interaction_level=random.choice(levels)
        meeting_mode_data=random.choice(meeting_mode)
        meeting_type_data=random.choice(meeting_type)
        host_location_data=random.choice(host_location)
        recording_data=random.choice(['yes','no'])
        
        df.loc[len(df)] =[organiser,call_duration_mins,participants,indiv_interaction_mins,
                          interaction_level,meeting_mode_data,meeting_type_data,host_location_data,recording_data]
        sys.stdout.write("\r{0}".format((float(i)/n)*100))
        sys.stdout.flush()
    return df.to_csv('skype_interactions_more.csv')
    #return df
```

```python
skype_interactions(10000)
```
